```python
x=np.linspace(1,10,10)#(start,end,how many value you wants)
y=x.reshape(2,5)
```

```python
y
```

```python
sns.heatmap(y,vmin=0,vmax=15,cmap="BrBG",annot=True)
plt.show()
```

```python
my_array=np.array([["a0","a1","a2","a3","a4"],["b0","b1","b2","b3","b4"]])
```

```python

sns.heatmap(y,vmin=0,vmax=15,cmap="Spectral",annot=my_array,fmt="s")
plt.show()
```

# Custom==> font,color
# Annot_Kws

```python
z={"fontsize":20,"color":"k"}
sns.heatmap(y,vmin=0,vmax=15,cmap="Spectral",annot=my_array,fmt="s",annot_kws=z)
plt.show()
```

# Line Width

```python
z={"fontsize":20,"color":"k"}
sns.heatmap(y,vmin=0,vmax=15,cmap="Spectral",annot=my_array,fmt="s",annot_kws=z,linewidths=10)
plt.show()
```

# Line Color

```python
z={"fontsize":20,"color":"k"}
sns.heatmap(y,vmin=0,vmax=15,cmap="Spectral",annot=my_array,fmt="s",annot_kws=z,
            linewidths=10,linecolor="r")
plt.show()
```

# Color Bar

```python
z={"fontsize":20,"color":"k"}
sns.heatmap(y,vmin=0,vmax=15,cmap="Spectral",annot=my_array,fmt="s",annot_kws=z,
            linewidths=10,linecolor="r",cbar=False)
plt.show()
```

# Xtick Labels,Ytick Labels

```python
z={"fontsize":20,"color":"k"}
sns.heatmap(y,vmin=0,vmax=15,cmap="Spectral",annot=my_array,fmt="s",annot_kws=z,
            linewidths=10,linecolor="r",cbar=False,xticklabels=False,yticklabels=False)
plt.show()
```

# Set Label

```python
z={"fontsize":20,"color":"k"}
d=sns.heatmap(y,vmin=0,vmax=15,cmap="Spectral",annot=my_array,fmt="s",annot_kws=z,
            linewidths=10,linecolor="r",cbar=False,xticklabels=False,yticklabels=False)

d.set(xlabel="Python",ylabel="Seaborn Heatmap")
plt.show()
```

# Change x,y Label font Scale Using Seaborn.set()

```python
z={"fontsize":20,"color":"k"}
d=sns.heatmap(y,vmin=0,vmax=15,cmap="Spectral",annot=my_array,fmt="s",annot_kws=z,
            linewidths=10,linecolor="r",cbar=False,xticklabels=False,yticklabels=False)

d.set(xlabel="Python",ylabel="Seaborn Heatmap")
sns.set(font_scale=3)
plt.show()
```
