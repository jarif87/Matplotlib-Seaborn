```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x=[1,2,3,4,5,6,7]
y=[1,2,3,5,8,9,10]
```

```python
df=pd.DataFrame({"first":x,"second":y})
df.head()
```

```python
sns.lineplot(x="first",y="second",data=df,marker="*",color="black")
plt.show()
```

```python
sns.lineplot(x=x,y=y)
plt.show()
```

```python
plt.plot(x,y)
plt.show()
```

```python
z=sns.load_dataset("penguins")
```

```python
z.head()
```

```python
plt.figure(figsize=(10,10))
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=z,hue="sex")
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=z,hue="sex",style="sex")
plt.show()
```

```python
z=sns.load_dataset("penguins").head(20)
```

```python
plt.figure(figsize=(10,10))
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=z,hue="sex",style="sex",palette="Accent",markers=["H","*"])
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=z,hue="sex",style="sex",palette="Accent",markers=["H","*"],dashes=False)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=z,hue="sex",style="sex",palette="Accent",markers=["H","*"],dashes=False,legend="full")
plt.show()
```
