```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

# Categorical scatterplots:

* stripplot() (with kind="strip"; the default)

* swarmplot() (with kind="swarm")

# Categorical distribution plots:

* boxplot() (with kind="box")

* violinplot() (with kind="violin")

* boxenplot() (with kind="boxen")

# Categorical estimate plots:

* pointplot() (with kind="point")

* barplot() (with kind="bar")

* countplot() (with kind="count")

```python
sns.catplot(x="size",y="tip",data=df,height=10)
plt.show()
```

```python
sns.catplot(x="size",y="tip",data=df,height=10,hue="sex")
plt.show()
```

```python
sns.catplot(x="size",y="tip",data=df,height=10,hue="sex",palette="PRGn")
plt.show()
```

```python
sns.catplot(x="day",y="size",data=df,height=10,hue="sex",palette="RdBu",kind="bar")
plt.show()
```

```python
sns.catplot(x="day",y="size",data=df,height=10,hue="sex",palette="afmhot",kind="point")
plt.show()
```

```python
sns.catplot(x="day",y="size",data=df,height=10,hue="sex",palette="cool",kind="box")
plt.show()
```

```python
sns.catplot(x="day",y="size",data=df,height=10,hue="sex",palette="gnuplot2",kind="boxen")
plt.show()
```
