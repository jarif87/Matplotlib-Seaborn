```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
y=[2,3,4,3,5,4]
```

```python
plt.stem(x,y)
plt.show()
```

```python
plt.stem(x,y,linefmt="--")
plt.show()
```

```python
plt.stem(x,y,linefmt="--",markerfmt="Pr")
plt.show()
```

```python
plt.stem(x,y,linefmt="--",markerfmt="Pr",bottom=4)
plt.show()
```

```python
plt.stem(x,y,linefmt="--",markerfmt="Pr",basefmt="magenta")
plt.show()
```

```python
plt.stem(x,y,linefmt="--",markerfmt="Pg",label="Python")
plt.legend()
plt.show()
```

```python
plt.stem(x,y,linefmt="--",markerfmt="Pg",label="Python",use_line_collection=False)
plt.legend()
plt.show()
```

```python
plt.stem(x,y,linefmt="--",markerfmt="Pg",label="Python",orientation="horizontal")
plt.legend()
plt.show()
```
