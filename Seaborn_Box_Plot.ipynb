```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
plt.figure(figsize=(10,10))
sns.set(style="darkgrid")
sns.boxplot(x="day",y="total_bill",data=df)
plt.show()
```

# Hue

```python
plt.figure(figsize=(10,10))
sns.set(style="darkgrid")
sns.boxplot(x="day",y="total_bill",data=df,hue="sex")
plt.show()
```

# Hue

```python
plt.figure(figsize=(10,10))
sns.set(style="darkgrid")
sns.boxplot(x="day",y="total_bill",data=df,hue="time")
plt.show()
```

# Color

```python
plt.figure(figsize=(10,10))
sns.set(style="darkgrid")
sns.boxplot(x="day",y="total_bill",data=df,hue="time",color="green")
plt.show()
```

# Order

```python
plt.figure(figsize=(10,10))
my_order=["Fri","Sat","Sun","Thur"]
sns.set(style="darkgrid")
sns.boxplot(x="day",y="total_bill",data=df,hue="time",order=my_order)
plt.show()
```

# Showmeans

```python
plt.figure(figsize=(10,10))
my_order=["Fri","Sat","Sun","Thur"]
sns.set(style="darkgrid")
sns.boxplot(x="day",y="total_bill",data=df,hue="time",showmeans=True)
plt.show()
```

# Meanprops

```python
mean_prop={"marker":"P","markeredgecolor":"navy","markerfacecolor":"darkorange","markersize":10}
plt.figure(figsize=(10,10))
my_order=["Fri","Sat","Sun","Thur"]
sns.set(style="darkgrid")
sns.boxplot(x="day",y="total_bill",data=df,hue="time",showmeans=True,meanprops=mean_prop)
plt.show()
```

# LineWidth

```python
mean_prop={"marker":"P","markeredgecolor":"navy","markerfacecolor":"darkorange","markersize":10}
plt.figure(figsize=(10,10))
my_order=["Fri","Sat","Sun","Thur"]
sns.set(style="darkgrid")
sns.boxplot(x="day",y="total_bill",data=df,hue="time",showmeans=True,meanprops=mean_prop,linewidth=4)
plt.show()
```

# Palette

```python
mean_prop={"marker":"P","markeredgecolor":"navy","markerfacecolor":"darkorange","markersize":10}
plt.figure(figsize=(10,10))
my_order=["Fri","Sat","Sun","Thur"]
sns.set(style="darkgrid")
sns.boxplot(x="day",y="total_bill",data=df,hue="time",showmeans=True,meanprops=mean_prop,linewidth=4,palette="gnuplot")
plt.show()
```

```python
plt.figure(figsize=(10,10))
my_order=["Fri","Sat","Sun","Thur"]
sns.set(style="darkgrid")
sns.boxplot(y="day",x="total_bill",data=df,hue="time",order=my_order)
plt.show()
```

# Orient

```python
plt.figure(figsize=(10,10))
my_order=["Fri","Sat","Sun","Thur"]
sns.set(style="darkgrid")
sns.boxplot(data=df,orient="h",palette="gist_rainbow_r")
plt.show()
```
