```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
x=["java","c","c++","python","R","c#"]
y=[30,34,50,40,30,60]
```

```python

plt.bar(x,y)

plt.xlabel("Language",fontsize=15)
plt.ylabel("Value",fontsize=15)
plt.title("Bar Plot\n",fontsize=20)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

color_name = ["red", "orange", "purple", "yellow", "navy", "green"]
z = [20, 30, 33, 50, 55, 60]

width = 0.2
x_array = np.arange(len(color_name))
remove_overlap = [val + width for val in z]

plt.figure(figsize=(10,10))
plt.barh(x_array, z, height=width, color="yellow", label="Languages_2")
plt.barh(x_array - width, y, height=width, color="red", label="Languages_1")
plt.xlabel("Value", fontsize=15)
plt.ylabel("Language", fontsize=15)
plt.title("Horizontal Bar Plot\n", fontsize=20)
plt.yticks(x_array, color_name, rotation=45, ha='right')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

color_dict = {"Java": "red", "C": "orange", "C++": "purple", "Python": "yellow", "R": "navy", "C#": "green"}

x = ["Java", "C", "C++", "Python", "R", "C#"]
y = [10, 8, 6, 4, 2, 1]

bars = plt.bar(x, y, color=[color_dict[label] for label in x])

plt.xlabel("Language", fontsize=15)
plt.ylabel("Value", fontsize=15)
plt.title("Bar Plot\n", fontsize=20)

handles = []
for label, color in color_dict.items():
    handles.append(plt.Rectangle((0,0),1,1, color=color))
plt.legend(handles, color_dict.keys())

plt.show()
```

```python
import matplotlib.pyplot as plt

color_dict = {"Java": "red", "C": "orange", "C++": "purple", "Python": "yellow", "R": "navy", "C#": "green"}

x = ["Java", "C", "C++", "Python", "R", "C#"]
y = [10, 8, 6, 4, 2, 1]

plt.figure(figsize=(10,10))
for i in range(len(x)):
    plt.bar(x[i], y[i], color=color_dict[x[i]], label=x[i])

plt.xlabel("Language", fontsize=15)
plt.ylabel("Value", fontsize=15)
plt.title("Bar Plot\n", fontsize=20)

plt.legend(loc="best")

plt.show()
```

```python
x=["python","c","c++","R","java"]
y=[20,30,40,44,33,55]
z=[40,50,33,44,20,60]
plt.figure(figsize=(10,10))
for i in range(len(x)):
  plt.bar(x[i],y[i],label=x[i],color="red")
  plt.bar(x[i],z[i],color="green")

plt.legend()
plt.show()
```
