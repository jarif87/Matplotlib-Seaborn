```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5]
y=[2,3,4,5,1]
plt.stackplot(x,y)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,3,4,5,1]
y1=[10,3,4,5,6]
y2=[4,5,6,8,9]
plt.stackplot(x,y,y1,y2)
plt.show()
```

```python
plt.figure(figsize=(10,10))
x=[1,2,3,4,5]
y=[2,3,4,5,1]
y1=[10,3,4,5,6]
y2=[4,5,6,8,9]
my_label=["y","y1","y2"]
plt.stackplot(x,y,y1,y2,labels=my_label)
plt.legend(loc="best",fontsize=15,labelcolor="navy")
plt.show()
```

```python
plt.figure(figsize=(10,10))
x=[1,2,3,4,5]
y=[2,3,4,5,1]
y1=[10,3,4,5,6]
y2=[4,5,6,8,9]
my_label=["y","y1","y2"]
my_color=["yellow","orange","blue"]
plt.stackplot(x,y,y1,y2,labels=my_label,colors=my_color)
plt.legend(loc="best",fontsize=15,labelcolor="navy")
plt.show()
```

```python
plt.figure(figsize=(10,10))
x=[1,2,3,4,5]
y=[2,3,4,5,1]
y1=[10,3,4,5,6]
y2=[4,5,6,8,9]
my_label=["y","y1","y2"]
my_color=["yellow","orange","blue"]
plt.stackplot(x,y,y1,y2,labels=my_label,colors=my_color,baseline="sym")
plt.legend(loc="best",fontsize=15,labelcolor="navy")
plt.show()
```

```python
plt.figure(figsize=(10,10))
x=[1,2,3,4,5]
y=[2,3,4,5,1]
y1=[10,3,4,5,6]
y2=[4,5,6,8,9]
my_label=["y","y1","y2"]
my_color=["yellow","orange","blue"]
plt.stackplot(x,y,y1,y2,labels=my_label,colors=my_color,baseline="wiggle")
plt.legend(loc="best",fontsize=15,labelcolor="navy")
plt.show()
```
