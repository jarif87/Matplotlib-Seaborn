```
import matplotlib.pyplot as plt
%matplotlib inline
```

```
x=["python","java","c++","c","R"]
y=[80,70,60,55,75]
plt.bar(x,y)
plt.show()
```

```
x=["python","java","c++","c","R"]
y=[80,70,60,55,75]
plt.bar(x,y)
plt.xlabel("Languages")
plt.ylabel("Number")
plt.title("Bar Plot\n",color="darkorange",fontsize=15)
plt.show()
```

```
x=["python","java","c++","c","R"]
y=[80,70,60,55,75]
plt.bar(x,y,width=0.4)
plt.xlabel("Languages")
plt.ylabel("Number")
plt.title("Bar Plot\n",color="darkorange",fontsize=15)
plt.show()
```

```
x=["python","java","c++","c","R"]
y=[80,70,60,55,75]
plt.bar(x,y,width=0.4,color="navy")
plt.xlabel("Languages")
plt.ylabel("Number")
plt.title("Bar Plot\n",color="darkorange",fontsize=15)
plt.show()
```

```
x=["python","java","c++","c","R"]
y=[80,70,60,55,75]
my_color=["green","blue","black","darkred","yellow","orange"]
plt.bar(x,y,width=0.4,color=my_color)
plt.xlabel("Languages")
plt.ylabel("Number")
plt.title("Bar Plot\n",color="darkorange",fontsize=15)
plt.show()
```

```
x=["python","java","c++","c","R"]
y=[80,70,60,55,75]
my_color=["green","blue","black","darkred","yellow","orange"]
plt.bar(x,y,width=0.4,color=my_color,align="edge")
plt.xlabel("Languages")
plt.ylabel("Number")
plt.title("Bar Plot\n",color="darkorange",fontsize=15)
plt.show()
```

```
x=["python","java","c++","c","R"]
y=[80,70,60,55,75]
my_color=["green","blue","black","darkred","yellow","orange"]
plt.bar(x,y,width=0.4,color=my_color,align="center")
plt.xlabel("Languages")
plt.ylabel("Number")
plt.title("Bar Plot\n",color="darkorange",fontsize=15)
plt.show()
```

```
x=["python","java","c++","c","R"]
y=[80,70,60,55,75]
my_color=["green","blue","black","darkred","yellow","orange"]
plt.bar(x,y,width=0.4,color=my_color,align="center",edgecolor="darkred")
plt.xlabel("Languages")
plt.ylabel("Number")
plt.title("Bar Plot\n",color="darkorange",fontsize=15)
plt.show()
```

```
x=["python","java","c++","c","R"]
y=[80,70,60,55,75]
my_color=["green","blue","black","darkred","yellow","orange"]
plt.bar(x,y,width=0.4,color=my_color,align="center",edgecolor="darkred",lw=4)
plt.xlabel("Languages")
plt.ylabel("Number")
plt.title("Bar Plot\n",color="darkorange",fontsize=15)
plt.show()
```

```
x=["python","java","c++","c","R"]
y=[80,70,60,55,75]
my_color=["green","blue","black","darkred","yellow","orange"]
plt.bar(x,y,width=0.4,color=my_color,align="center",edgecolor="darkred",lw=4,linestyle="--")
plt.xlabel("Languages")
plt.ylabel("Number")
plt.title("Bar Plot\n",color="darkorange",fontsize=15)
plt.show()
```

```
x=["python","java","c++","c","R"]
y=[80,70,60,55,75]
my_color=["green","blue","black","darkred","yellow","orange"]
plt.bar(x,y,width=0.4,color=my_color,align="center",alpha=0.4)
plt.xlabel("Languages")
plt.ylabel("Number")
plt.title("Bar Plot\n",color="darkorange",fontsize=15)
plt.show()
```

```
x=["python","java","c++","c","R"]
y=[80,70,60,55,75]
my_color=["green","blue","black","darkred","yellow","orange"]
plt.bar(x,y,width=0.4,color=my_color,label="Language Popularity")
plt.xlabel("Languages")
plt.ylabel("Number")
plt.title("Bar Plot\n",color="darkorange",fontsize=15)
plt.legend(fontsize=14)
plt.show()
```
