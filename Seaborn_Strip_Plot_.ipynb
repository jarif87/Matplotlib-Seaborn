```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
plt.figure(figsize=(10,10))
sns.stripplot(x="day",y="total_bill",data=df)
plt.show()
```

# Hue

```python
plt.figure(figsize=(10,10))
sns.stripplot(x="day",y="total_bill",data=df,hue="sex")
plt.show()
```

# Patlette

```python
plt.figure(figsize=(10,10))
sns.stripplot(x="day",y="total_bill",data=df,hue="sex",palette="copper_r")
plt.show()
```

# Line Width

```python
plt.figure(figsize=(10,10))
sns.stripplot(x="day",y="total_bill",data=df,hue="sex",linewidth=2)
plt.show()
```

# Edge Color

```python
plt.figure(figsize=(10,10))
sns.stripplot(x="day",y="total_bill",data=df,hue="sex",linewidth=2,edgecolor="g")
plt.show()
```

# Jitter

```python
plt.figure(figsize=(10,10))
sns.stripplot(x="day",y="total_bill",data=df,hue="sex",jitter=3)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.stripplot(x="day",y="total_bill",data=df,hue="sex",size=10)
plt.show()
```

# Marker

```python
plt.figure(figsize=(10,10))

sns.stripplot(x="day", y="total_bill", hue="sex", data=df, marker="P",size=12)
plt.show()
```

# Alpha

```python
plt.figure(figsize=(10,10))

sns.stripplot(x="day", y="total_bill", hue="sex", data=df, marker="P",size=12,alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(10,10))

sns.stripplot(x=df["total_bill"])
plt.show()
```

```python
plt.figure(figsize=(10,10))

sns.stripplot(y=df["total_bill"],marker="*",size=10)
plt.show()
```
