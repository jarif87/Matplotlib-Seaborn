```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[1,2,3,4,5,6,7]
plt.plot(x,y)
plt.fill_between(x=[2,4],y1=2,y2=4)
plt.title("Fill Between Plot")
plt.show()
```

```python
x=[1,2,3,4,5,6,7]
y=[1,2,3,4,5,6,7]
plt.figure(figsize=(10,10))
plt.plot(x,y,color="darkorange",marker="H",lw=4)
plt.fill_between(x,y)
plt.title("Fill Between Plot")
plt.show()
```

```python
plt.figure(figsize=(10,10))
x=[1,2,3,4,5,6,7]
y=[1,2,3,4,5,6,7]
plt.plot(x,y)
plt.fill_between(x=[5,6,7],y1=5,y2=7)
plt.title("Fill Between Plot")
plt.show()
```

```python
x=[1,2,3,4,5,6,7]
y=[1,2,3,4,5,6,7]
plt.figure(figsize=(10,10))
plt.plot(x,y,color="darkorange",marker="H",lw=4)
plt.fill_between(x,y,color="navy")
plt.title("Fill Between Plot")
plt.show()
```

```python
x=np.array([1,2,3,4,5,6,7])
y=np.array([1,2,3,4,5,6,7])
plt.figure(figsize=(10,10))
plt.plot(x,y,color="darkorange",marker="H",lw=4)
plt.fill_between(x,y,color="navy",where=(x>=2) & (x<=4))
plt.title("Fill Between Plot")
plt.show()
```
