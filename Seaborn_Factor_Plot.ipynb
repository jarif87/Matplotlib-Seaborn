```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
sns.factorplot(x="size",y="tip",data=df)
plt.show()
```

```python

sns.factorplot(x="size",y="tip",data=df,hue="sex",height=10)
plt.show()
```

```python
sns.factorplot(x="size",y="tip",data=df,hue="sex",height=10,kind="bar")
plt.show()
```

```python
sns.factorplot(x="size",y="tip",data=df,hue="sex",height=10,kind="box")
plt.show()
```
