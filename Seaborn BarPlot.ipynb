```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset("penguins")
```

```python
df.head()
```

# Seaborn Barplot

```python
sns.barplot(x="island",y="bill_length_mm",data=df)
plt.show()
```

# Matplotlib Barplot

```python
c = ['red', 'yellow', 'black']
plt.bar(df["island"],df["bill_length_mm"])
```

# Hue

```python
sns.barplot(x="island",y="bill_length_mm",data=df,hue="sex")
plt.show()
```

# Order

```python
order=["Dream","Biscoe","Torgersen"]
sns.barplot(x="island",y="bill_length_mm",data=df,hue="sex",order=order)
plt.show()
```

# Hue Order

```python
order=["Dream","Biscoe","Torgersen"]
hue_order=["Female","Male"]
sns.barplot(x="island",y="bill_length_mm",data=df,hue="sex",order=order,hue_order=hue_order)
plt.show()
```

# Ci

```python
order=["Dream","Biscoe","Torgersen"]
hue_order=["Female","Male"]
sns.barplot(x="island",y="bill_length_mm",data=df,hue="sex",order=order,hue_order=hue_order,errorbar=('ci', 80))
plt.show()
```

# N Boot

```python
order=["Dream","Biscoe","Torgersen"]
hue_order=["Female","Male"]
sns.barplot(x="island",y="bill_length_mm",data=df,hue="sex",order=order,hue_order=hue_order,errorbar=('ci', 80),n_boot=3)
plt.show()
```

# Palette

```python
order=["Dream","Biscoe","Torgersen"]
hue_order=["Female","Male"]
sns.barplot(x="island",y="bill_length_mm",data=df,palette="RdBu",saturation=5)
plt.show()
```

# Errcolor ,Errwidth

```python
sns.barplot(x="island",y="bill_length_mm",data=df,palette="RdBu",errcolor="red",errwidth=5)
plt.show()
```

# Capsize

```python
sns.barplot(x="island",y="bill_length_mm",data=df,errcolor="red",errwidth=5,capsize=1)
plt.show()
```

# Dodge

```python
sns.barplot(x="island",y="bill_length_mm",data=df,hue="sex",dodge=False)
plt.show()
```

# Alpha

# Darkgrid

```python
sns.set(style="darkgrid")
sns.barplot(x="island",y="bill_length_mm",data=df,hue="sex",saturation=300,alpha=0.8)
plt.show()
```
