```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

# Violinplot

```python
sns.violinplot(x="day",y="total_bill",data=df)
plt.show()
```

# Hue

```python
sns.violinplot(x="day",y="total_bill",data=df,hue="time")
plt.show()
```

# Linewidth

```python
sns.violinplot(x="day",y="total_bill",data=df,hue="time",linewidth=3)
plt.show()
```

# Palette

```python
sns.violinplot(x="day",y="total_bill",data=df,palette="Dark2")
plt.show()
```

# Color

```python
sns.violinplot(x="day",y="total_bill",data=df,color="green")
plt.show()
```

```python
df.head()
```

# Order

```python
sns.violinplot(x="smoker",y="tip",data=df,order=["Yes","No"],palette="Paired")
plt.show()
```

```python
sns.violinplot(x="sex",y="total_bill",data=df,order=["Male","Female"],palette="Set3_r")
plt.show()
```

```python
sns.violinplot(x="sex",y="total_bill",data=df,palette="Set3_r",saturation=0.8)
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=df,split=True,hue="sex",scale="width")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=df,split=True,hue="sex",scale="area")
plt.show()
```

# Inner

```python
sns.violinplot(x="sex",y="total_bill",data=df,order=["Male","Female"],palette="Set3_r",inner="quart")
plt.show()
```

```python
sns.violinplot(x="sex",y="total_bill",data=df,order=["Male","Female"],palette="Set3_r",inner="point")
plt.show()
```

```python
sns.violinplot(x="sex",y="total_bill",data=df,order=["Male","Female"],palette="Set3_r",inner="stick")
plt.show()
```
