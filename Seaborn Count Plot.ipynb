```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

# Load DataSet

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
sns.countplot(x="sex",data=df)
plt.show()
```

```python
sns.barplot(x="sex",y="size",data=df)
plt.show()
```

```python
sns.countplot(x="sex",data=df,hue="smoker")
plt.show()
```

```python
sns.countplot(x="sex",data=df,hue="smoker",palette="viridis")
plt.show()
```

```python
sns.countplot(x="sex",data=df,hue="smoker",color="lime")
plt.show()
```

```python
sns.countplot(x="sex",data=df,hue="smoker",color="lime",saturation=0.2)
plt.show()
```
