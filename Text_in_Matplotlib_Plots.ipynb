```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# plot

```python
x=[1,2,3,4,5]
y=[2,3,4,5,1]
plt.figure(figsize=(10,10))
plt.plot(x,y)
plt.show()
```

# title

```python
x=[1,2,3,4,5]
y=[2,3,4,5,1]
plt.figure(figsize=(10,10))
plt.plot(x,y)
plt.title("Text In Plot\n",color="navy",fontsize=15)
plt.show()
```

# xlabel,ylabel

```python
x=[1,2,3,4,5]
y=[2,3,4,5,1]
plt.figure(figsize=(10,10))
plt.plot(x,y)
plt.title("Text In Plot\n",color="navy",fontsize=15)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

# plt.text

```python
x=[1,2,3,4,5]
y=[2,3,4,5,1]
plt.figure(figsize=(10,10))
plt.plot(x,y)
plt.title("Text In Plot\n",color="navy",fontsize=15)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.text(3.5,5,"Python",fontsize=15,color="darkorange",style="italic")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,3,4,5,1]
plt.figure(figsize=(10,10))
plt.plot(x,y)
plt.title("Text In Plot\n",color="navy",fontsize=15)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.text(3.5,5,"Python",fontsize=15,color="darkorange",style="italic",bbox={"facecolor":"darkblue"})
plt.show()
```

# plt.annotate

```python
x=[1,2,3,4,5]
y=[2,3,4,5,1]
plt.figure(figsize=(10,10))
plt.plot(x,y)
plt.title("Text In Plot\n",color="navy",fontsize=15)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.text(3.5,5,"Python",fontsize=15,color="darkorange",style="italic",bbox={"facecolor":"darkblue"})
plt.annotate("my_point",xy=(3,2),xytext=(3,3.5),arrowprops=dict(facecolor="black",shrink=400))
plt.show()
```

# plt.legend

```python
x=[1,2,3,4,5]
y=[2,3,4,5,1]
plt.figure(figsize=(10,10))
plt.plot(x,y)
plt.title("Text In Plot\n",color="navy",fontsize=15)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(["my_legend"],loc=7) # value=(0 to 10)
plt.show()
```

# plt.legend (more parameters)

```python
x=[1,2,3,4,5]
y=[2,3,4,5,1]
plt.figure(figsize=(10,10))
plt.plot(x,y)
plt.title("Text In Plot\n",color="navy",fontsize=15)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(["my_legend"],facecolor="brown",fontsize=20) # value=(0 to 10)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,3,4,5,1]
plt.figure(figsize=(10,10))
plt.plot(x,y)
plt.title("Text In Plot\n",color="navy",fontsize=15)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(["my_legend"],facecolor="brown",fontsize=10,edgecolor="darkred") # value=(0 to 10)
plt.show()
```
