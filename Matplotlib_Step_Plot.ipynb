```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[6,4,5,3,2,1,8]
plt.step(x,y)
plt.show()
```

```python
x=[1,2,3,4,5,6,7]
y=[6,4,5,3,2,1,8]
plt.step(x,y)
plt.grid(True)
plt.show()
```

```python
x=[1,2,3,4,5,6,7]
y=[6,4,5,3,2,1,8]
plt.step(x,y)
plt.grid(True)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
x=[1,2,3,4,5,6,7]
y=[6,4,5,3,2,1,8]
plt.step(x,y)
plt.grid(True)
plt.title("Step Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
plt.figure(figsize=(10,10))
x=[1,2,3,4,5,6,7]
y=[6,4,5,3,2,1,8]
plt.step(x,y,marker="*")
plt.grid(True)
plt.title("Step Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
plt.figure(figsize=(10,10))
x=[1,2,3,4,5,6,7]
y=[6,4,5,3,2,1,8]
plt.step(x,y,color="navy",marker="H",markersize=10,markerfacecolor="darkorange")
plt.grid(True)
plt.title("Step Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
plt.figure(figsize=(10,10))
x=[1,2,3,4,5,6,7]
y=[6,4,5,3,2,1,8]
plt.step(x,y,color="navy",marker="H",markersize=10,markerfacecolor="darkorange",label="Step Plot")
plt.grid(True)
plt.title("Step Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(loc=2,fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(10,10))
x=[1,2,3,4,5,6,7]
y=[6,4,5,3,2,1,8]
plt.step(x,y,color="navy",marker="H",markersize=10,markerfacecolor="darkorange",label="Step Plot")
plt.grid(True)
plt.title("Step Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(loc=2,fontsize=15,labelcolor="blue")
plt.show()
```
