```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[10,20,30,40,50]
y=["c","c++","java","R","python"]
```

# pie chart

```python
plt.pie(x)
plt.show()
```

# labels

```python
plt.pie(x,labels=y)
plt.show()
```

# explode

```python
ex=[0.0,0.0,0.2,0.3,0.5]
plt.pie(x,labels=y,explode=ex)
plt.show()
```

# colors

```python
ex=[0.0,0.0,0.2,0.3,0.5]
my_color=["red","green","navy","black","yellow"]
plt.pie(x,labels=y,explode=ex,colors=my_color)
plt.show()
```

# autopct

```python
ex=[0.0,0.0,0.2,0.3,0.5]
my_color=["red","green","navy","orange","yellow"]
plt.pie(x,labels=y,explode=ex,colors=my_color,autopct="%0.2f%%")
plt.show()
```

# shadow

```python

my_color=["red","green","navy","orange","yellow"]
plt.pie(x,labels=y,colors=my_color,autopct="%0.2f%%",shadow=True)
plt.show()
```

# radius

```python
my_color=["red","green","navy","orange","yellow"]
plt.pie(x,labels=y,colors=my_color,autopct="%0.2f%%",shadow=True,radius=2)
plt.show()
```

# labeldistance

```python
my_color=["red","green","navy","orange","yellow"]
plt.pie(x,labels=y,colors=my_color,autopct="%0.2f%%",radius=1.5,labeldistance=0.8)
plt.show()
```

```python
my_color=["red","green","navy","orange","yellow"]
plt.pie(x,labels=y,colors=my_color,autopct="%0.2f%%",radius=1.5,labeldistance=1.2)
plt.show()
```

# start angle

```python
my_color=["red","green","navy","orange","yellow"]
plt.pie(x,labels=y,colors=my_color,autopct="%0.2f%%",radius=1.5,startangle=90)
plt.show()
```

# textprops

```python
my_color=["red","green","navy","orange","yellow"]
plt.pie(x,labels=y,colors=my_color,autopct="%0.2f%%",radius=1.5,textprops={"fontsize":18,"color":"black"})
plt.show()
```

# counter clock

```python
my_color=["red","green","navy","orange","yellow"]
ex=[0.0,0.0,0.1,0.2,0.3]
plt.pie(x,labels=y,colors=my_color,explode=ex,autopct="%0.2f%%",radius=1.5,counterclock=False)
plt.show()
```

# wedgeprops

```python
my_color=["red","green","navy","orange","yellow"]
ex=[0.0,0.0,0.0,0.0,0.3]

plt.pie(x,labels=y,colors=my_color,explode=ex,autopct="%0.2f%%",radius=1.5,wedgeprops={"linewidth":10,"edgecolor":"m"})
plt.show()
```

# Center

```python
my_color=["red","green","navy","orange","yellow"]
ex=[0.0,0.0,0.0,0.0,0.3]

plt.pie(x,labels=y,colors=my_color,explode=ex,autopct="%0.2f%%",radius=1.5,
        wedgeprops={"linewidth":10,"edgecolor":"m"},center=(8,8))
plt.show()
```

# rotate labels

```python
my_color=["red","green","navy","orange","yellow"]
ex=[0.0,0.0,0.0,0.0,0.3]

plt.pie(x,labels=y,colors=my_color,explode=ex,autopct="%0.2f%%",radius=1.5,rotatelabels=True)
plt.show()
```

```python
my_color=["red","green","navy","orange","yellow"]
ex=[0.0,0.0,0.0,0.0,0.3]

plt.pie(x,labels=y,colors=my_color,explode=ex,autopct="%0.2f%%",radius=1.5,rotatelabels=True)
plt.legend(loc=2)
plt.show()
```

```python
plt.pie([1])
plt.show()
```

```python
z=[50,60,20,30,40]
plt.pie(x,labels=y,radius=1.0)
plt.pie(z,radius=0.4)
plt.show()
```

```python
z=[50,60,20,30,40]
sec_color=["green","darkorange","darkred","darkblue","purple"]
plt.pie(x,labels=y,radius=1.0)
plt.pie(z,radius=0.5,colors=sec_color)
plt.show()
```

# Donut Pie Chart

```python
plt.pie(x,labels=y,radius=1.5)
plt.pie([1],colors="white")
plt.show()
```

```python
plt.pie(x,labels=y,radius=1.5)
my_circle=plt.Circle(xy=(0,0),radius=1,facecolor="white")
plt.gca().add_artist(my_circle)
plt.show()
```
