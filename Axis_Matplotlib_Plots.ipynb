```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,6,4,5,7]
y=[4,5,6,7,2,1,3,6]
plt.style.use("seaborn-notebook")
plt.figure(figsize=(10,10))
plt.plot(x,y)

plt.show()
```

```python
x=[1,2,3,4,6,4,5,7]
y=[4,5,6,7,2,1,3,6]
plt.style.use("seaborn-dark")
plt.figure(figsize=(10,10))
plt.plot(x,y)
plt.xticks(x)
plt.yticks(y)
plt.show()
```

```python
x=[1,2,3,4,6,4,5,7]
y=[4,5,6,7,2,1,3,6]
plt.style.use("seaborn-dark-palette")
plt.figure(figsize=(10,10))
my_label=["C","C++","Java","Python","Php","R","Javascript","Ruby"]
plt.plot(x,y)
plt.xticks(x,labels=my_label)
plt.yticks(y,labels=my_label)
plt.show()
```

```python
x=[1,2,3,4,6,4,5,7]
y=[4,5,6,7,2,1,3,6]
plt.style.use("classic")
plt.figure(figsize=(10,10))
plt.plot(x,y)

plt.xlim(0,10)
plt.ylim(0,10)
plt.show()
```

```python
x=[1,2,3,4,6,4,5,7]
y=[4,5,6,7,2,1,3,6]
plt.style.use("_classic_test_patch")
plt.figure(figsize=(10,10))
plt.plot(x,y)
plt.axis([0,10,0,8])
plt.show()
```
