```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[2,3,5,4,1,8,6]
z=[4,25,60,7,8,11,20]
```

# scatter plot

```python
plt.scatter(x,y)
plt.title("ScatterPlot\n",fontsize=15,color="darkorange")
plt.xlabel("Day",fontsize=10,color="darkorange")
plt.ylabel("Number",fontsize=10,color="darkorange")
plt.show()
```

# color

# figsize

```python
plt.figure(figsize=(10,10))
plt.scatter(x,y,color="red")
plt.title("ScatterPlot\n",fontsize=15,color="darkorange")
plt.xlabel("Day",fontsize=10,color="darkorange")
plt.ylabel("Number",fontsize=10,color="darkorange")
plt.show()
```

# C for Each Color

```python
plt.figure(figsize=(10,10))
colors=["g","k","r","c","y","b","m"]
for i in range(len(x)):
  plt.scatter(x[i],y[i],color=colors[i])

plt.title("ScatterPlot\n",fontsize=15)
plt.xlabel("Day",fontsize=10,color="darkorange")
plt.ylabel("Number",fontsize=10,color="darkorange")
plt.show()
```

# sizes for each value

```python
plt.figure(figsize=(10,10))
sizes=[10,20,30,40,50,60,70]
plt.scatter(x,y,c=colors,s=sizes)
plt.title("ScatterPlot\n",fontsize=15,color="darkorange")
plt.xlabel("Day",fontsize=10,color="darkorange")
plt.ylabel("Number",fontsize=10,color="darkorange")
plt.show()
```

# different color for each value another method

```python
plt.figure(figsize=(10,10))
sizes=[100,200,300,400,500,600,700]
plt.scatter(x,y,c=colors,s=sizes)
plt.title("ScatterPlot\n",fontsize=15,color="darkorange")
plt.xlabel("Day",fontsize=10,color="darkorange")
plt.ylabel("Number",fontsize=10,color="darkorange")
plt.show()
```

# alpha

```python
plt.figure(figsize=(10,10))
sizes=[100,200,300,400,500,600,700]
plt.scatter(x,y,c=colors,s=sizes,alpha=0.4)
plt.title("ScatterPlot\n",fontsize=15,color="darkorange")
plt.xlabel("Day",fontsize=10,color="darkorange")
plt.ylabel("Number",fontsize=10,color="darkorange")
plt.show()
```

# set grid style

```python
plt.figure(figsize=(10,10))
plt.style.use("seaborn-colorblind")
sizes=[100,200,300,400,500,600,700]
plt.scatter(x,y,c=colors,s=sizes)
plt.title("ScatterPlot\n",fontsize=15,color="darkorange")
plt.xlabel("Day",fontsize=10,color="darkorange")
plt.ylabel("Number",fontsize=10,color="darkorange")
plt.show()
```

# marker

```python
plt.figure(figsize=(10,10))
sizes=[100,200,300,400,500,600,700]
plt.scatter(x,y,c=colors,s=sizes,marker="*")
plt.title("ScatterPlot\n",fontsize=15,color="darkorange")
plt.xlabel("Day",fontsize=10,color="darkorange")
plt.ylabel("Number",fontsize=10,color="darkorange")
plt.show()
```

# marker edge color

```python
plt.figure(figsize=(10,10))
sizes=[100,200,300,400,500,600,700]
plt.scatter(x,y,c=colors,s=sizes,marker="*",edgecolor="navy")
plt.title("ScatterPlot\n",fontsize=15,color="darkorange")
plt.xlabel("Day",fontsize=10,color="darkorange")
plt.ylabel("Number",fontsize=10,color="darkorange")
plt.show()
```

# linewidth

```python
plt.figure(figsize=(10,10))
sizes=[100,200,300,400,500,600,700]
plt.scatter(x,y,c=colors,s=sizes,marker="*",edgecolor="navy",lw=4)
plt.title("ScatterPlot\n",fontsize=15,color="darkorange")
plt.xlabel("Day",fontsize=10,color="darkorange")
plt.ylabel("Number",fontsize=10,color="darkorange")
plt.show()
```

# Color Bar

```python
plt.figure(figsize=(10,10))
sizes=[100,200,300,400,500,600,700]
my_color=[10,30,40,60,70,80,20] #color range=0-100
plt.scatter(x,y,c=my_color,s=sizes,cmap="inferno")
color_bar=plt.colorbar()
color_bar.set_label("Color_Bar",fontsize=15)
plt.title("ScatterPlot\n",fontsize=15,color="darkorange")
plt.xlabel("Day",fontsize=10,color="darkorange")
plt.ylabel("Number",fontsize=10,color="darkorange")
plt.show()
```

# multiple plot

```python
plt.scatter(x,y,color="green",s=sizes)
plt.scatter(x,z,color="red",s=sizes)
plt.title("ScatterPlot\n",fontsize=15,color="darkorange")
plt.xlabel("Day",fontsize=10,color="darkorange")
plt.ylabel("Number",fontsize=10,color="darkorange")
plt.show()
```
