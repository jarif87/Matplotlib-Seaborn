```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

# Read Data

```python
df=sns.load_dataset("penguins")
```

```python
df.head()
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=df)
plt.show()
```

# Hue

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=df,hue="sex")
plt.show()
```

# style,sizes

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=df,hue="sex",
                sizes=(150,50),style="sex")
plt.show()
```

# size

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=df,hue="sex",
                sizes=(100,50),style="sex",size="sex")
plt.show()
```

# palette

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=df,hue="sex",
                sizes=(100,50),style="sex",size="sex",palette="bwr_r")
plt.show()
```

# alpha

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=df,hue="sex",
                sizes=(100,50),style="sex",size="sex",palette="bwr_r",alpha=0.2)
plt.show()
```

# marker

```python
my_marker={"Male":"*","Female":"o"}
plt.figure(figsize=(10,10))
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=df,hue="sex",style="sex",markers=my_marker,size=100)
plt.show()
```
