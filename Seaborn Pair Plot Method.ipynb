```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

# Pairplot

```python
sns.pairplot(df)
plt.show()
```

# Hue

```python
sns.pairplot(df,hue="sex")
plt.show()
```

# Vars

```python
sns.pairplot(df,vars=["total_bill","tip"],hue="sex")
plt.show()
```

# Hue Order

```python
sns.pairplot(df,vars=["total_bill","tip"],hue="sex",hue_order=["Female","Male"])
plt.show()
```

# Paletter

```python
sns.pairplot(df,hue="sex",palette="GnBu")
plt.show()
```

# X_Vars

```python
sns.pairplot(df,hue="sex",palette="plasma",x_vars=["size","sex","tip"])
plt.show()
```

# Y_Vars

```python
sns.pairplot(df,hue="sex",palette="autumn",y_vars=["total_bill","sex","smoker"])
plt.show()
```

# Kind

### Kind="scatter","kde","hist","reg"

```python
sns.pairplot(df,hue="sex",palette="autumn",kind="reg")
plt.show()
```

```python
sns.pairplot(df,hue="sex",palette="cool",kind="scatter")
plt.show()
```

```python
sns.pairplot(df,hue="sex",palette="summer",kind="hist")
plt.show()
```

```python
sns.pairplot(df,hue="sex",palette="Wistia",kind="kde")
plt.show()
```

# Diag_Kind

```python
sns.pairplot(df,hue="sex",palette="winter",kind="kde",diag_kind="hist")
plt.show()
```

```python
sns.pairplot(df,hue="sex",palette="winter",kind="kde",diag_kind="kde")
plt.show()
```

```python
sns.pairplot(df,hue="sex",palette="hot",markers=["*","P"])
plt.show()
```
