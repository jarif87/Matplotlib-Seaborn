```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
x
```

```python
y=x*3
```

```python
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,color="green",linestyle=":",linewidth=3)
plt.title("First Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
y1=x*3
y2=x*4
```

```python
plt.plot(x,y1,color="darkorange",linestyle=":",linewidth=3)
plt.plot(x,y2,color="darkred",linestyle=":",linewidth=4)
plt.title("First Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()
```

# Adding Subplots

```python
plt.subplot(1,2,1)
plt.plot(x,y1,color="blue",linestyle=":",linewidth=3)
plt.grid(True)
plt.subplot(1,2,2)
plt.plot(x,y2,color="purple",linestyle=":",linewidth=3)
plt.grid(True)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1,color="blue",linestyle=":",linewidth=3)
plt.grid(True)
plt.subplot(2,1,2)
plt.plot(x,y2,color="purple",linestyle=":",linewidth=3)
plt.grid(True)
plt.show()
```

# Barplot

```python
dict0={"anika":40,"fariha":60,"urmi":60,"rakul":75,"mardia":80,"tamanna":90}
```

```python
name=list(dict0.keys())
value=list(dict0.values())
```

```python
plt.bar(name,value)
plt.title("Bar Plot")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

# Barh

```python
plt.barh(name,value,color="lime")
plt.title("Bar Plot")
plt.ylabel("Names")
plt.xlabel("Marks")
plt.grid(True)
plt.show()
```

# Scatter Plot

```python
x=[1,2,3,4,5,6,7,8,9]
y=[4,3,6,7,8,9,10,1,2]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y,color="yellow",marker="*",s=100)
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[4,3,6,7,8,9,10,1,2]
z=[15,44,60,79,55,17,11,20,47]
```

```python
plt.scatter(x,y,color="red",marker="*",s=100)
plt.scatter(x,z,color="indigo",marker=".",s=200)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(x,y,color="red",marker="*",s=100)
plt.grid(True)
plt.subplot(1,2,2)
plt.scatter(x,z,color="indigo",marker=".",s=200)
plt.grid(True)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.scatter(x,y,color="red",marker="*",s=100)
plt.grid(True)
plt.subplot(2,1,2)
plt.scatter(x,z,color="indigo",marker=".",s=200)
plt.grid(True)
plt.show()
```
